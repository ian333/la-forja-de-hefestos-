Mostramos primero los contornos con **PEOR error** para entender los patrones de fallo.

```python
from PIL import Image
from IPython.display import display
import os, json

JPG_DIR = '/home/ian/Projects/GAIA/la-forja-de-hefestos-/debug-jpg'
DATA_FILE = '/home/ian/Projects/GAIA/la-forja-de-hefestos-/debug-report/data.json'

with open(DATA_FILE) as f:
    data = json.load(f)

data.sort(key=lambda c: c['maxErr'], reverse=True)

files = sorted([f for f in os.listdir(JPG_DIR) if f.endswith('.jpg')])
idx_to_file = {}
for f in files:
    parts = f.split('_')
    idx = int(parts[1])
    idx_to_file[idx] = f

print(f"Total contornos: {len(data)}")
print(f"Con maxErr > 10: {sum(1 for c in data if c['maxErr'] > 10)}")
print(f"Con maxErr > 1:  {sum(1 for c in data if c['maxErr'] > 1)}")
print(f"OK (maxErr < 1): {sum(1 for c in data if c['maxErr'] <= 1)}")
```
